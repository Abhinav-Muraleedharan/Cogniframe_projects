```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
import matplotlib.pyplot as plt
```

```python
# Assuming X_train_sample is your DataFrame
# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_sample)

# Apply PCA
pca = PCA()
pca.fit(X_train_scaled)

# Plotting the Cumulative Summation of the Explained Variance
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(pca.explained_variance_) + 1), pca.explained_variance_, marker='o', linestyle='--')
plt.title('Explained Variance by Each Component - Eigenvalue Decay')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.grid(True)
plt.show()
```

```python
# Calculate the cumulative variance
cumulative_variance = pca.explained_variance_ratio_.cumsum()

# Plotting the Cumulative Summation of the Explained Variance
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='-')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```
